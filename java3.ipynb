Q1. Here's a simple example of a banking system program using object-oriented programming concepts in Java:

```java
class BankAccount {
    private String accountHolderName;
    private double balance;

    public BankAccount(String accountHolderName, double initialBalance) {
        this.accountHolderName = accountHolderName;
        this.balance = initialBalance;
    }

    public String getAccountHolderName() {
        return accountHolderName;
    }

    public double getBalance() {
        return balance;
    }

    public void deposit(double amount) {
        balance += amount;
        System.out.println("Amount deposited: " + amount);
    }

    public void withdraw(double amount) {
        if (balance >= amount) {
            balance -= amount;
            System.out.println("Amount withdrawn: " + amount);
        } else {
            System.out.println("Insufficient balance");
        }
    }
}

public class BankingSystem {
    public static void main(String[] args) {
        BankAccount account = new BankAccount("John Doe", 1000.0);

        System.out.println("Account Holder Name: " + account.getAccountHolderName());
        System.out.println("Current Balance: " + account.getBalance());

        account.deposit(500.0);
        account.withdraw(200.0);

        System.out.println("Updated Balance: " + account.getBalance());
    }
}
```

In this program, the `BankAccount` class represents a bank account. It has private variables for the account holder's name and the account balance. The class provides methods to retrieve the account holder's name, balance, and perform operations like deposit and withdrawal.

The `BankingSystem` class serves as the entry point of the program. It creates an instance of `BankAccount`, sets initial values, and demonstrates the usage of various methods.

Q2. Inheritance and method overriding can be demonstrated as follows:

```java
class Parent {
    public void display() {
        System.out.println("Parent's display method");
    }
}

class Child extends Parent {
    @Override
    public void display() {
        System.out.println("Child's display method");
    }
}

public class MethodOverridingExample {
    public static void main(String[] args) {
        Parent parent = new Parent();
        parent.display(); // Output: Parent's display method

        Child child = new Child();
        child.display(); // Output: Child's display method

        Parent polymorphicChild = new Child();
        polymorphicChild.display(); // Output: Child's display method (runtime polymorphism)
    }
}
```

In this example, the `Parent` class has a `display()` method. The `Child` class extends `Parent` and overrides the `display()` method with its own implementation. In the `main()` method, we create instances of both `Parent` and `Child` classes and call the `display()` method on them. The output shows that when we call the `display()` method on the parent reference, it executes the overridden method in the child class (runtime polymorphism).

Q3. Runtime polymorphism (also known as dynamic polymorphism) can be demonstrated as follows:

```java
class Animal {
    public void makeSound() {
        System.out.println("Animal is making a sound");
    }
}

class Dog extends Animal {
    @Override
    public void makeSound() {
        System.out.println("Dog is barking");
    }
}

class Cat extends Animal {
    @Override
    public void makeSound() {
        System.out.println("Cat is meowing");
    }
}

public class RuntimePolymorphismExample {
    public static void main(String[] args) {
        Animal animal1 = new Dog();
        Animal animal2 = new Cat();

        animal1.makeSound(); // Output: Dog is barking
        animal2.makeSound(); // Output: Cat is meowing
    }
}
```

In this example, the `Animal` class has a `makeSound()` method. The `Dog` and `Cat` classes extend `Animal` and override the `makeSound()` method with their own implementations. In the `main()` method, we create instances of both `Dog` and `Cat` classes, but they are assigned to the `Animal` reference variables. When we call the `makeSound()` method on these references, the actual implementation of the respective subclasses is invoked (runtime polymorphism).

Q4. Compile-time polymorphism (also known as static polymorphism) can be achieved through method overloading. Here's an example:

```java
class Calculator {
    public int add(int a, int b) {
        return a + b;
    }

    public double add(double a, double b) {
        return a + b;
    }

    public String add(String a, String b) {
        return a + b;
    }
}

public class CompileTimePolymorphismExample {
    public static void main(String[] args) {
        Calculator calculator = new Calculator();

        int result1 = calculator.add(5, 10);
        double result2 = calculator.add(2.5, 3.7);
        String result3 = calculator.add("Hello", " World");

        System.out.println("Result 1: " + result1); // Output: 15
        System.out.println("Result 2: " + result2); // Output: 6.2
        System.out.println("Result 3: " + result3); // Output: Hello World
    }
}
```

In this example, the `Calculator` class has multiple `add()` methods with different parameter types. Depending on the argument types provided, the appropriate `add()` method is selected at compile-time. The program demonstrates how the `add()` method can accept integers, doubles, and strings and perform the addition accordingly.

Q5. Achieving loose coupling in Java using object-oriented programming concepts can be done through encapsulation and dependency inversion. Here's a simplified explanation:

- **Encapsulation:** Encapsulation is a concept that hides the internal implementation details of a class and provides a well-defined interface for interacting with it. It allows objects to communicate with each other through well-defined methods, while the internal state remains hidden. By encapsulating the implementation details, changes to one class do not directly affect other classes that depend on it. This promotes loose coupling between classes.

- **Dependency Inversion:** Dependency inversion is a principle that promotes loose coupling by abstracting dependencies and relying on interfaces or abstract classes instead of concrete implementations. By programming to interfaces, classes can depend on abstractions rather than specific implementations. This allows for flexibility and interchangeable dependencies, making it easier to substitute or modify dependencies without affecting the overall system.

By following these principles, you can reduce direct dependencies between classes and create a modular and flexible system. This promotes code reusability, maintainability, and easier testing and debugging.

Q6. Encapsulation in Java provides several benefits:
- Data Hiding: Encapsulation hides the internal details of a class and allows access to data only through well-defined methods. This protects the data from being modified or accessed incorrectly, ensuring data integrity and security.
- Modularity: Encapsulation promotes modularity by organizing code into separate units (classes) with their own data and methods. This makes the code easier to understand, maintain, and modify.
- Code Reusability: Encapsulation allows for code reusability as classes can be used as building blocks in different parts of a program without exposing their internal implementation.
- Flexibility: By encapsulating implementation details, changes can be made to a class without affecting other parts of the program that depend on it. This promotes flexibility and reduces the risk of introducing bugs or breaking existing code.

Q7. Java is often considered a "pure" object-oriented programming language, but it is not 100% object-oriented due to the following reasons:
- Primitive Data Types: Java includes primitive data types (int, char, boolean, etc.) that are not objects and do not inherit from a common class. However, Java provides wrapper classes (Integer, Character, Boolean, etc.) to treat these primitives as objects when needed.
- Static Members: Java allows the declaration of static methods and variables that belong to the class itself, rather than an instance of the class. These static members do not rely on object instantiation and are not considered object-oriented.
- Procedural Programming: Java supports procedural programming constructs, such as procedural methods, which can be used outside the context of objects and classes.
- Final Classes: Java has final classes that cannot be subclassed, which deviates from the core principle of inheritance in object-oriented programming.

Despite these deviations, Java predominantly follows object-oriented principles and supports key features such as encapsulation, inheritance, and polymorphism.

Q8. Abstraction in Java provides several advantages:
- Simplifies Complexity: Abstraction simplifies complex systems by breaking them down into manageable modules. It allows programmers to focus on essential details and hide unnecessary complexity, making code more readable and maintainable.
- Encapsulation of Implementation: Abstraction allows for the encapsulation of implementation details within classes, exposing only relevant information through well-defined interfaces. This promotes data hiding and reduces dependencies on internal implementation details.
- Code Reusability: Abstraction promotes code reusability by creating abstract classes or interfaces that can be implemented by multiple classes. This allows for the creation of interchangeable components and facilitates code sharing and modularity.
- Flexibility and Adaptability: Abstraction enables the creation of flexible and adaptable systems. By programming to interfaces or abstract classes, implementations can be easily swapped or extended without affecting the overall system.

Q9. Abstraction is a concept that allows us to represent essential features and behaviors while hiding unnecessary details. It focuses on the "what" rather than the "how." An example of abstraction can be seen in a car:

```java
interface Vehicle {
    void start();
    void stop();
}

class Car implements Vehicle {
    @Override
    public void start() {
        System.out.println("Car started");
    }

    @Override
    public void stop() {
        System.out.println("Car stopped");
    }
}

public class AbstractionExample {
    public static void main(String[] args) {
        Vehicle car = new Car();
        car.start();
        car.stop();
    }
}
```

In this example, the `Vehicle` interface represents a generic vehicle and defines the behaviors of starting and stopping. The `Car` class implements the `Vehicle` interface and provides the specific implementation for starting and stopping a car. The `main()` method demonstrates how the abstracted behavior of starting and stopping is called on a `Car` object without worrying about the internal details of the implementation.

Q10. In Java, a final class is a class that cannot be subclassed or extended further. It is the last level in the inheritance hierarchy. When a class is declared as final, it means that its implementation is complete, and it cannot be modified or overridden by any other class. Some reasons to use a final class are:

- Security: By making a class final, you prevent other classes from subclassing it and potentially overriding or modifying sensitive behavior or data.
- Performance: The compiler can perform optimizations on final classes since their behavior is fixed, resulting in improved performance.
- Design Decisions: Sometimes, a class is intentionally made final to enforce design decisions or prevent unintended modifications to its behavior.

For example, the `String` class in Java is a final class. This means that you cannot create a subclass of `String` and modify its fundamental behavior. It ensures that the core functionality and immutability of strings are maintained throughout the Java ecosystem.
